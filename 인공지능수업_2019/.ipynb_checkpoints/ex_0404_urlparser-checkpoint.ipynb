```python
import urllib.request
from bs4 import BeautifulSoup 

url = "https://finance.naver.com/marketindex/"
response = urllib.request.urlopen(url)

soup = BeautifulSoup(response, "html.parser")
results= soup.select("span.value")
for result in results:
    print(result.string)
```

```python
 # 로그인 과정 + 데이터 수집 
    
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

USER= "h960305"
PASS= "Happy0305"
session = requests.session()

login_info={
    "m_id":USER,
    "m_passwd": PASS
}
url_login="http://www.hanbit.co.kr/member/login_proc.php"

# post방식으로 로그인 ( 보안 강화된 경우 이 방법 이용 불가)
res=session.post(url_login, data = login_info)
res.raise_for_status() # 오류 시 예외 발생 

#마이페이지 접근 
url_mypage= "http://www.hanbit.co.kr/myhanbit/myhanbit.html"
res=session.get(url_mypage)
res.raise_for_status()

#res는 HTML 객체이므로 res.text를 사용하여 parsing
soup = BeautifulSoup(res.text, "html.parser")
mileage=soup.select_one(".mileage_section1").get_text()
ecoin = soup.select_one(".mileage_section2").get_text()

print("마일리지: "+mileage)
print("이코인: "+ecoin)
```

```python
import urllib.request as request
import json 

json_str=request.urlopen("https://api.github.com/repositories").read()
output = json.loads(json_str)

for item in output:
    print(item["name"])
    print(item["full_name"])
    print(item["owner"]["login"])
```

```python

# codecs 라이브러리 활용하여 euc_kr  디코딩 
import codecs 
# euc_kr로 저장된 csv파일 읽기
filename="./data/test.csv"
csv= codecs.open(filename,"r","euc_kr").read()
# csv를 파이썬 리스트로 변환
data =[]
rows= csv.split("\r\n")
for row in rows:
    if row=="": continue
    cells = row.split(",")
    data.append(cells)
for c in data:
    print(c[0],c[1],c[2])
```

```python
# csv모듈 이용해 csv 파일 

import csv 

filename= "./data/test.csv"
with open(filename,"r") as f:
    reader = csv.reader(f, delimiter=",", quotechar='"')
    print(reader)
    
    for cells in reader:
        print(cells[0],cells[1],cells[2])
```
